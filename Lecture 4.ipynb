```python
for link in soup('a', {'class': 'storylink'}):
    print(link['href'])
```

```python
import urllib.parse
```

```python
urllib.parse.urlparse('http://www.bbc.com/news/technology-40492668')
```

```python
def get_link_domains():
    for link in soup('a', {'class': 'storylink'}):
        yield urllib.parse.urlparse(link['href']).netloc
```

```python
from collections import Counter
```

```python
domain_counts = Counter(get_link_domains())
```

```python
domain_counts.most_common(3)
```
